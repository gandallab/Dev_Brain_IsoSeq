````r
library("tidyverse")
library("ggplot2")
library("rtracklayer")
````

````r
options(repr.plot.width = 6, repr.plot.height = 4, repr.plot.res = 300)
````

````r
gtf = rtracklayer::import("cp_vz_0.75_min_7_recovery_talon.gtf") %>% as_tibble()
gtf
````

````r
transcripts = gtf %>% filter(type == "transcript") %>% select(where(~!all(is.na(.x))))
transcripts
````

````r
colnames(transcripts)
````

````r
n_distinct(transcripts$transcript_id)
````

````r
transcripts = transcripts %>% mutate(novelty = case_when(
    transcript_status == "KNOWN" ~ "Known",
    ISM_transcript == "TRUE" ~ "ISM",
    genomic_transcript == "TRUE" ~ "Genomic",
    NIC_transcript == "TRUE" ~ "NIC",
    NNC_transcript == "TRUE" ~ "NNC",
    antisense_transcript == "TRUE" ~ "Antisense",
    intergenic_transcript == "TRUE" ~ "Intergenic"
) %>% fct_infreq() %>% fct_relevel("Known"))
````

````r
unique(transcripts$novelty)
````

````r
transcripts = transcripts %>%
    mutate(novelty = novelty %>% fct_other(drop = c("Antisense", "Genomic", "Intergenic")))
````

````r
colors = c(
  "Known" = "#009E73",
  "ISM"   = "#0072B2",
  "NIC"   = "#D55E00",
  "NNC"   = "#E69F00",
  "Other" = "#000000"
)
````

````r
iso_types = transcripts %>%
    count(novelty) %>%
    mutate(freq = n/sum(n)) %>%
    mutate(prop = n / sum(n) * 100) %>%
    mutate(prop_formatted = sprintf("%.1f%%", prop))

ggplot(iso_types, aes(x = novelty, y = n, fill=novelty)) +
    geom_col() +
    geom_text(
        aes(label = prop_formatted),
        vjust = c(1.5, 1.5, 1.5, 1.5, -0.2),
        colour = c("white", "white", "white", "white", "black")
    ) +
    scale_y_continuous(labels = scales::comma) +
    scale_fill_manual(values = colors) +
    guides(fill = "none") +
    xlab("Category") +
    ylab("Number of transcripts")
````

````r
exons = gtf %>%
    filter(type == "exon") %>%
    left_join(select(transcripts, transcript_id, novelty), by = "transcript_id")
````

````r
tx_lengths = exons %>%
  group_by(transcript_id, novelty) %>%
  summarize(length = sum(width)) %>%
  mutate(novelty_simple = novelty %>% fct_other(keep = "Known", other_level = "Novel")) %>%
  drop_na()
````

````r
ggplot(tx_lengths, aes(x = length, fill = novelty)) +
  geom_histogram(bins = 75) +
  scale_x_continuous(limits = c(0, 11000)) +
  coord_cartesian(xlim = c(0, 10000)) +
  labs(
    x = "Transcript length (bp)",
    y = "Count",
    fill = "Category"
  ) +
  scale_fill_manual(
    values = colors,
    breaks = c("Known", "ISM", "NIC", "NNC", "Other")
  ) + 
  theme_gray(base_size = 14) + 
  theme(
    legend.position = c(.954545454545455, .9321562368),
    legend.justification = c(1, 1),
    axis.text = element_text(color="black")
  )
````

````r
levels = c(
  "Protein coding" = "protein_coding",
  "lncRNA" = "lncRNA",
  {
    x = str_subset(unique(gtf$gene_type), "pseudogene")
    names(x) = rep("Pseudogene", each = length(x))
    x
  },
  "Novel gene" = "Novel gene",
  "lincRNA" = "lincRNA"
)

gene_types = gtf %>%
  filter(type == "gene") %>%
  select(gene_id, gene_type) %>%
  mutate(
    gene_type = gene_type %>%
      replace_na("Novel gene") %>%
      fct_recode(!!!levels) %>%
      fct_other(keep = unique(names(levels))) %>%
      fct_relevel(unique(names(levels)))
  )
````

````r
ggplot(gene_types, aes(y = fct_rev(gene_type))) +
  geom_bar(aes(x = (..count..)/sum(..count..), fill = gene_type)) +
  scale_x_continuous(labels = scales::label_percent()) +
  guides(fill = "none") +
  labs(
    x = "Percentage",
    y = "Gene type"
  ) +
  theme_gray(base_size = 14)
````
